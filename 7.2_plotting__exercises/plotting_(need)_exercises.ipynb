```python
import matplotlib.pyplot as plt

# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='darkred', linestyle='--',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with error')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np

def func(x):
    return np.exp(-x/10) * np.sin(x)

x = np.linspace(0, 50, 1000)
y = func(x)

params = {
    'fontsize': 15,
    'color': 'darkorange',
    'rotation': 0
}

plt.plot(x, y)
plt.xlabel('x', params)
plt.ylabel('y', params)
plt.ylim([-1, 1])
plt.title(r'$y=e^{-x/10}sin(x)$', fontsize = 20)
plt.grid(alpha=.25)
plt.show()
```
