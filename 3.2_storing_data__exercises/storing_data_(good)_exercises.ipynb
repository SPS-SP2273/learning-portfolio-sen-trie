# Exercise 2

```python
import numpy as np

np_array_2d = np.array([[1, "A"], [3, "C"], [2, "B"], [4, "D"],
                        [5, "E"], [7, "G"], [6, "F"], [8, "H"],
                        [10, "J"], [9, "I"]])

print(np_array_2d[np_array_2d[:, 0].astype(int) % 2 == 1][:, 1])
```

# Exercise 3

```python
np_array = np.array([[66, 50, 57, -1, -1],
 [92, -1, 88, -1, -1],
 [75, -1, -1, 76, -1],
 [-1, 51, 87, -1, -1],
 [71, -1, 67, -1, 69]])

np_array[np_array < 50] = -1
np_array[np_array > 50] = 1
np_array[np_array == 50] = 0

print(np_array)
```
