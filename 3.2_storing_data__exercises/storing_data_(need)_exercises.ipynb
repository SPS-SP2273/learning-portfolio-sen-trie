```python
x_1 = [1, 4, 9, 16, 25, 36, 49, 64, 81, 100]
x_2 = [1, 8, 27, 64, 125, 216, 343, 512, 729, 1000]

x_1_mean=np.mean(x_1)
x_2_mean=np.mean(x_2)

x_1_std=np.std(x_1)
x_2_std=np.std(x_2)

top = np.sum((np.array(x_1) - x_1_mean) * (np.array(x_2) - x_2_mean))
bot = np.sqrt(np.sum((np.array(x_1) - x_1_mean)**2) * np.sum((np.array(x_2) - x_2_mean)**2))

print(f"x_1_mean: {x_1_mean}, x_1_mean: {x_2_mean}")
print(f"x_1_std: {x_1_std}, x_2_std: {x_2_std}")
print(top/bot)
print(np.corrcoef(x_1,x_2)[0, 1])
```
