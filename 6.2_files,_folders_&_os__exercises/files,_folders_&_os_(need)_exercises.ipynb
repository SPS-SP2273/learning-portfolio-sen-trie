```python
mon_num = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}

def rename_my_file(old_file_name):
    name_split = old_file_name.split("-")
    return (f"{mon_num[name_split[0]]}-{name_split[1]}")

file_list = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']

for file in file_list:
    print(f"{file} ----> {rename_my_file(file)}")
```

```python
import glob
import os

for files in glob.glob('os-collaboration-exercise-data/*.txt'):
    path = files.split('/')[0]
    name = files.split('/')[1]
    
    os.rename(f"{path}/{name}", f"{path}/{rename_my_file(name)}")
```

```python
cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for city in cities:
    try:
        os.mkdir(f'os-collaboration-exercise-data/{city}')
    except FileExistsError:
        print(f"Directory already exists, skipping")
    except Exception as e:
        print(f"An error occurred: {e}")

```

```python
# Task 6

import shutil

for files in glob.glob('os-collaboration-exercise-data/*.txt'):
    path = files.split('/')[0]
    name = files.split('/')[1]

    if 'Cairo'.lower() in name.lower():
        shutil.copy(files, f"{path}/Cairo")
```

```python
# Task 7

for files in glob.glob('os-collaboration-exercise-data/*.txt'):
    path = files.split('/')[0]
    name = files.split('/')[1]
    city = name.split('_')[1].split('.')[0]

    shutil.copy(files, f"{path}/{city.capitalize()}")
```
