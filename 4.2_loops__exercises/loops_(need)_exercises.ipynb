```python
for num in range(1, 6):
    print(f'{num} :', end=' ')
    for mul in range(1, 6):
        print(f'{str(num * mul).ljust(4)}', end='')
    print('')
```

```python
import numpy as np

for i in range(10):
    exp = np.random.rand(10)
    heads = len(exp[exp > 0.5])
    print(f"Experiment {str(i + 1).rjust(2)}: No. of Heads = {heads}")
```
