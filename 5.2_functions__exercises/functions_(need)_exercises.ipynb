'if' statements do not require an 'else' statement to execute correctly

```python
def calculate_bmi(weight, height):
    bmi = weight / (height ** 2)
    if bmi < 18.5:
        return 'Underweight'
    elif bmi < 24.9:
        return 'Normal weight'
    elif bmi < 29.9:
        return 'Overweight'
    else:
        return 'Obese'

print(calculate_bmi(59, 1.70))    
```

```python
def factorial(n):
    if n < 0:
        raise ValueError("Negative numbers are not allowed")
    elif n == 0:
        return 1    
    else:
        return n * factorial(n-1)

print(factorial(31))
```
